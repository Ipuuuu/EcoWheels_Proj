```python
# Check your dataset stats
import os
from collections import Counter

# Count images per class
train_dir = '../data/train/labels'
val_dir = '../data/validation/labels'


def count_classes(dir_path):
    counts = Counter()
    for file in os.listdir(dir_path):
        if file.endswith('.txt'):
            with open(os.path.join(dir_path, file)) as f:
                for line in f:
                    class_id = int(line.split()[0])
                    counts[class_id] += 1
    return counts

print('Training distribution:', count_classes(train_dir))
print('Validation distribution:', count_classes(val_dir))
```

## Optimized Training Parameters for Class Imbalance:


```python

# ============================================================
# HELPER FUNCTION: Calculate class weights for imbalanced data
# ============================================================

def calculate_class_weights(class_distribution, method='sqrt'):
    """
    Calculate class weights to handle imbalance
    
    Args:
        class_distribution: dict of {class_id: count}
        method: 'inverse', 'sqrt', or 'log'
    
    Returns:
        list of weights for each class
    """
    import numpy as np
    
    total_samples = sum(class_distribution.values())
    n_classes = len(class_distribution)
    
    weights = []
    for class_id in sorted(class_distribution.keys()):
        count = class_distribution[class_id]
        
        if method == 'inverse':
            # Inverse frequency
            weight = total_samples / (n_classes * count)
        elif method == 'sqrt':
            # Square root inverse (less extreme)
            weight = np.sqrt(total_samples / (n_classes * count))
        elif method == 'log':
            # Logarithmic (even less extreme)
            weight = np.log(total_samples / count + 1)
        
        weights.append(weight)
    
    # Normalize to have mean of 1.0
    weights = np.array(weights)
    weights = weights / weights.mean()
    
    return weights.tolist()


# Example usage:
train_dist = {59: 503, 36: 333, 58: 321, 29: 206, 39: 191, 5: 176, 
              12: 152, 7: 133, 9: 130, 55: 98, 57: 79, 14: 76, 6: 75, 
              21: 73, 50: 71, 8: 62, 27: 58, 17: 55, 33: 51, 0: 49, 
              40: 43, 20: 42, 16: 35, 4: 33, 31: 31, 10: 29, 42: 27, 
              38: 25, 45: 24, 34: 21, 18: 21, 51: 18, 49: 18, 15: 9, 
              22: 9, 46: 9, 52: 8, 28: 8, 25: 7, 11: 7, 43: 7, 54: 6, 
              2: 6, 53: 5, 47: 5, 30: 5, 32: 5, 37: 4, 26: 4, 44: 3, 
              56: 3, 48: 3, 41: 3, 23: 3, 19: 2, 13: 2, 1: 2, 24: 1}

class_weights = calculate_class_weights(train_dist, method='sqrt')
print("\nCalculated class weights (first 10):", class_weights[:10])
print("To use these, add to your data.yaml:")
print(f"class_weights: {class_weights}")
```

```python
from ultralytics import YOLO
import torch

# Load model
model = YOLO('yolo11s.pt')

# Train with optimized parameters for class imbalance
results = model.train(
    data="../yolo_taco/data.yaml",
    
    # ============ TRAINING DURATION ============
    epochs=150,              # Increased from 100 - imbalanced data needs more time
    patience=25,             # Increased from 15 - prevent premature stopping
    
    # ============ IMAGE & BATCH SETTINGS ============
    imgsz=640,               # Changed from 416 - better for small objects (trash items)
    batch=8,                 # Slightly increased from 6 (you have VRAM)
    workers=4,               # Increased from 2 for faster data loading
    
    # ============ DEVICE & PRECISION ============
    device=0,
    amp=True,
    half=True,
    
    # ============ LEARNING RATE ============
    lr0=0.001,               # Reduced from 0.01 - start more conservative
    lrf=0.01,                # Final LR will be lr0 * lrf
    cos_lr=True,
    warmup_epochs=5.0,       # Added - gradual LR warmup
    warmup_momentum=0.8,     # Added
    
    # ============ OPTIMIZER ============
    optimizer='AdamW',       # Better than SGD for imbalanced data
    weight_decay=0.0005,
    momentum=0.937,
    
    # ============ DATA AUGMENTATION (Critical for imbalance!) ============
    # Geometric augmentations
    degrees=15.0,            # Increased from default 0.0 - rotation
    translate=0.2,           # Increased from default 0.1 - translation
    scale=0.9,               # Increased from default 0.5 - scaling
    shear=5.0,               # Added - shear transformation
    perspective=0.0005,      # Added - perspective transformation
    flipud=0.0,              # No vertical flip (trash orientation matters)
    fliplr=0.5,              # Horizontal flip
    
    # Color augmentations
    hsv_h=0.02,              # Increased - hue variation
    hsv_s=0.7,               # Saturation variation
    hsv_v=0.4,               # Value/brightness variation
    
    # Advanced augmentations
    mosaic=1.0,              # Keep mosaic - great for small objects
    mixup=0.15,              # Added - blend images (helps minority classes)
    copy_paste=0.3,          # Added - copy-paste augmentation (critical for rare classes!)
    
    # ============ LOSS WEIGHTS (Address class imbalance) ============
    box=7.5,                 # Default is 7.5
    cls=0.5,                 # Reduced from default - less penalty on classification
    dfl=1.5,                 # Default is 1.5
    
    # ============ CLASS WEIGHTS ============
    # NOTE: You'll need to calculate these based on your distribution
    # See the helper function below
    
    # ============ OTHER SETTINGS ============
    save=True,
    save_period=10,          # Save checkpoint every 10 epochs
    exist_ok=True,
    name="yolo_taco_optimized",
    pretrained=True,
    verbose=True,
    seed=42,
    
    # ============ VALIDATION ============
    val=True,
    plots=True,              # Generate training plots
    
    # ============ EARLY STOPPING & CHECKPOINTING ============
    save_hybrid=False,       # Don't save hybrid ONNX models
    rect=False,              # Don't use rectangular training (can hurt small batch sizes)
    resume=False,            # Set to True if continuing training
    
    # ============ ADVANCED: Multi-scale training ============
    # Uncomment if you want to train at multiple scales (slower but more robust)
    # multi_scale=True,
)

print("\n" + "="*60)
print("TRAINING COMPLETED!")
print(f"Best model: runs/detect/{results.save_dir}/weights/best.pt")
print(f"Last model: runs/detect/{results.save_dir}/weights/last.pt")
print("="*60)

```

```python
import cv2
import torch
import numpy as np
from ultralytics import YOLO

# Load your trained model
model = YOLO('/home/immaculatapatrickumoh/Documents/EcoWheels_Proj/runs/detect/train2/weights/best.pt')  # Your trained model


# URL formats:
# - DroidCam: http://192.168.5.115:4747/video
# - DroidCam OBS: might be different, check OBS settings

# IP = "192.168.5.117"  # Your phone's IP address
# PORT = "4747"  # Default DroidCam port
# URL = f"http://{IP}:{PORT}/video"
URL = f"https://ai-project-nine-sable.vercel.app/"


def run_inference_on_camera():
    # Open the video stream
    cap = cv2.VideoCapture(URL)
   
    if not cap.isOpened():
        print(f"Error: Could not open video stream from {URL}")
        print("Make sure:")
        print("1. DroidCam is running on your phone")
        print("2. Your PC and phone are on same WiFi")
        print("3. IP address is correct")
        return
   
    print(f"Connected to {URL}")
    print("Press 'q' to quit")
   
    while True:
        # Read frame
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break
       
        # Run YOLO inference
        results = model(frame, verbose=False)
       
        # Draw detections
        annotated_frame = results[0].plot()
       
        # Display FPS
        cv2.putText(annotated_frame, f"FPS: {int(cap.get(cv2.CAP_PROP_FPS))}",
                    (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
       
        # Show the frame
        cv2.imshow('Phone Camera - YOLO Detection', annotated_frame)
       
        # Break loop on 'q' press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
   
    # Cleanup
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    run_inference_on_camera()
```
